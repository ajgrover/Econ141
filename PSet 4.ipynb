```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as sm
```

```python
dfs = pd.read_excel("hprice.xls", sheet_name=None)
data = dfs['hprice1']
```

```python
price_reg = sm.ols(formula="price ~ sqrft + bdrms", data=data).fit()
print(price_reg.summary())
```

```python
price_reg_no_bed = sm.ols(formula="price ~ sqrft", data=data).fit()
print(price_reg_no_bed.summary())
```

```python
price_reg_w_lot = sm.ols(formula="price ~ sqrft + bdrms + lotsize", data=data).fit()
print(price_reg_w_lot.summary())
```
