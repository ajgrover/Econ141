```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as sm
```

```python
dfs = pd.read_excel("cps98_short.xls", sheet_name=None)
data = dfs['cps98_short']
```

```python
hourly_on_age_gend_edu = sm.ols(formula="ahe ~ age + female + bachelor", data=data).fit()
print(hourly_on_age_gend_edu.summary())
```

```python
data['ln_ahe'] = data['ahe'].apply(np.log)
log_hourly_on_age_gend_edu = sm.ols(formula="ln_ahe ~ age + female + bachelor", data=data).fit()
print(log_hourly_on_age_gend_edu.summary())
```

```python
data['ln_ahe'] = data['ahe'].apply(np.log)
data['ln_age'] = data['age'].apply(np.log)
log_hourly_on_log_age_gend_edu = sm.ols(formula="ln_ahe ~ ln_age + female + bachelor", data=data).fit()
print(log_hourly_on_log_age_gend_edu.summary())
```

```python
part_d_males = data[data['female'] == 0]
part_d_females = data[data['female'] == 1]
part_d_males_reg = sm.ols(formula="ahe ~ age", data=part_d_males).fit()
print('males', part_d_males_reg.summary())
part_d_females_reg = sm.ols(formula="ahe ~ age", data=part_d_females).fit()
print('females', part_d_females_reg.summary())
```
